```python
#***important stuff in this cell are Center of Mass Energy (CME) and s.***


#Energy of electron = E_e = 18 GeV
#Energy of proton = E_p = 275 GeV
#center of Mass Energy = CME = sqrt(s)
E_e = 18
E_p = 275
s = 4*E_e*E_p
CME = np.sqrt(s)
print(CME)

#Measure of Inelasticity = y
#E_e of scattered electron = E_e' = E_es
#E_es = []

Y = [y*0.01 for y in range(10, 96)]
#print(Y)
E = [eta*0.01 for eta in range(-177, 132)]
#print(E)

for y in Y:
    for eta in E:
        theta = 2*np.arctan(np.exp(-eta))
    E_es = ((-E_e*(y-1))/(np.cos(theta/2)**2))
    #print(E_es)

for eta in E:
    theta = 2*np.arctan(np.exp(-eta))
    y = 1 - (E_e/E_es)*(np.cos(theta/2))**2
    #print(y)
    if y > 0.1 and y < 0.95:
        plt.scatter(theta, y)
    #print(E_e,E_es)
    #print(theta)

#o-plus = sqrt(___^2 + ___^2)
```

```python
#Resolution power
#squared momentum transfer of virtual photon = Q^2 = MT
Theta = []
E_s = []
Q = []
X = []
Xp = []
Xm = []
Qp = []
Qm = []
Eta = []
Ys = []
X1 = []
Q1 = []
X95 = []
Q95 = []
Y2 = []
Q2 = []
Etam = []
Etap = []


E = [eta*0.01 for eta in range(-400, 401)]
Y = [y*0.01 for y in range(10, 96)]
for eta in E:
    theta = 2*np.arctan(np.exp(-eta))
    #print(theta)
    for y in Y:
        #print(Y)
        if y >= 0.1 and y <= 0.96:
            E_es = ((-E_e*(y-1))/(np.cos(np.pi/2 - theta/2)**2))
            MT = 2*E_e*E_es*(1-np.cos(np.pi - theta))
            x = MT/(s*y)
            if eta > 0 and eta <= 1.31:
                Xp.append(x)
                Qp.append(MT)
                Etap.append(eta)
            if eta <= 0 and eta >= -1.77:
                Xm.append(x)
                Qm.append(MT)
                Etam.append(eta)
            if y == 0.1:
                X1.append(x)
                Q1.append(MT)
            if y == 0.9500000000000001:
                X95.append(x)
                Q95.append(MT)
            y2 = (x*s)/MT
            q2 = s*y*x
            Eta.append(eta)
            Theta.append(theta)
            E_s.append(E_es)
            Q.append(MT)
            X.append(x)
            Ys.append(y)
            Y2.append(y2)
            Q2.append(q2)
            #print(eta, y, MT, x)
#print(Eta)
#print(Theta)
#print(E_s)
#print(X)
#print(Q)
#print(Qm, Xm)
#print(Qp, Xp)
#print(X1,Q1)
#print(X95,Q95)
#print(Ys)

#fig = plt.figure(figsize = (20,12))
#ax =fig.add_subplot(2,2,1)
#ay = fig.add_subplot(2,2,2)
#az = fig.add_subplot(2,2,3)
#aw = fig.add_subplot(2,2,4)

plt.scatter(X, Q, c='black')
plt.scatter(Xm, Qm, c='purple')
plt.scatter(Xp, Qp, c='red')
plt.plot(X1, Q1, lw=2.5)
plt.plot(X95, Q95, lw=2.5)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Bjorken x")
plt.ylabel("$Q^2$ ($GeV^2$)")
plt.title("$Q^2$ vs Bjorken x at 18 x 275 GeV")
plt.legend(['Full Calorimeter range: $\eta$ = [-4.00,4.00]','ECal Backscatter range: $\eta$ = [-1.77,0]','ECal Frontscatter range: $\eta$ = [0,1.31]','y=0.1','y=0.95'], loc='upper left', fontsize='small')
plt.show()

plt.savefig('Q^2_vs_Bjorken_x_plot.png')

#plt.scatter(Theta, E_s)
#plt.xscale("log")
#plt.yscale("log")
#plt.xlabel("Scattering Angle of Electron ($\theta$)")
#plt.ylabel("Scattered Electron Energy ($E_s$)")

#plt.scatter(X, Y2)
#plt.xscale("log")
#plt.yscale("log")
#plt.xlabel("Bjorken x")
#plt.ylabel("Inelasticity")

#plt.scatter(Ys, Q2)
#plt.xscale("log")
#plt.yscale("log")
#plt.xlabel("Inelasticity")
#plt.ylabel("$Q^2$")

#ax.legend()
```
